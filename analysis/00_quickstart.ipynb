# 00 – Quick Start

A two-minute tour of the **dv/v coupling framework**. No external data or API keys required.

By the end you will have:
1. computed a coda **sensitivity depth** with one function call,
2. built and validated a **site configuration**, and
3. loaded a **named preset** from `presets/`.

> Run this notebook with the working directory set to `analysis/` (the default when opened via `pixi run lab`).

## 1. One number: how deep does the coda sample?

`sensitivity_depth` applies the third-wavelength rule, $d \approx \tfrac{1}{3}\,V_S/f$, to estimate the depth most sensitive to a coda-wave dv/v measurement.

```python
from poroelastic_framework import sensitivity_depth

# 2 Hz coda in granite (Vs = 2500 m/s)
depth_m = sensitivity_depth(Vs_mps=2500, freq_hz=2.0)
print(f"Sensitivity depth: {depth_m:.1f} m")
```

## 2. Build and validate a configuration

`SiteConfig` / `AnalysisConfig` are immutable, validated schemas. Out-of-range
values (e.g. a Poisson ratio ≥ 0.5) raise a `ValidationError` immediately.

```python
from config import PARKFIELD, AnalysisConfig, validate_and_summarize

cfg = AnalysisConfig(site=PARKFIELD, frequency_hz=3.0)
print(validate_and_summarize(cfg))
```

```python
# Validation catches unphysical inputs before any computation runs
from config import SiteConfig
from pydantic import ValidationError

try:
    SiteConfig(name="bad", Vs=2500, rho=2700, nu=0.7, mu_prime=10)
except ValidationError as exc:
    print("Rejected as expected:\n", exc)
```

## 3. Load a named preset

Reusable recipes live under `presets/`. Each bundles a site, analysis settings,
a description, and a literature reference.

```python
from config import load_analysis_config

cfg = load_analysis_config("../presets/cascadia_subduction.yaml")
print(validate_and_summarize(cfg))
```
